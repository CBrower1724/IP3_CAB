### Question 3 part A

```python
X = np.linspace(0, 100, 100)
X_squared = X ** 2
fig, ax1 = plt.subplots()
ax1.plot(X, X_squared, label='X vs X^2', color='red')
ax1.set_title("X vs X^2")
ax1.set_xlabel("X")
ax1.set_ylabel("X^2")
plt.show()
```

### Question 3 part B

```python
fig, ax2 = plt.subplots()
ax2.scatter(X, np.exp(X), label='X vs exponential(X)', color='blue', s=10)  # s for marker size
ax2.set_title("X vs exponential(X)")
ax2.set_xlabel("X")
ax2.set_ylabel("exponential(X)")
plt.show()
```

### Question 4

```python
car_sales=pd.read_csv("car-sales (1).csv")
car_sales.info()
```

### Question 5

```python
car_sales['Price'] = car_sales['Price'].replace({'\$': '', ',': ''}, regex=True).astype('int64')
car_sales.head()
```

### Question 6

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

### Question 7

```python
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp('today'), periods=len(car_sales)).date
car_sales.head()
```

### Question 8

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title('Price vs Total Sales line plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

### Question 9

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'], color='purple', s=100, alpha=0.7)
plt.title('Odometer vs Price scatter plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

### Question 10

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'], color='skyblue')
plt.title('Make vs Odometer bar plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

### Question 11

```python
sns.histplot(car_sales['Make'], bins=4)
plt.title("Histogram of car make")
```

### Question 12

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
axs[0, 0].plot(car_sales['Price'], car_sales['Total Sales'])
axs[0,0].set_title('Price vs Total Sales line plot')
axs[0,0].set_xlabel('Price')
axs[0,0].set_ylabel('Total Sales')

axs[0, 1].scatter(car_sales['Odometer (KM)'], car_sales['Price'], color='purple', s=100, alpha=0.7)
axs[0,1].set_title('Odometer vs Price scatter plot')
axs[0,1].set_xlabel('Odometer (KM)')
axs[0,1].set_ylabel('Price')

axs[1, 0].bar(car_sales['Make'], car_sales['Odometer (KM)'], color='skyblue')
axs[1,0].set_title('Make vs Odometer bar plot')
axs[1,0].set_xlabel('Make')
axs[1,0].set_ylabel('Odometer (KM)')

sns.countplot(x=car_sales['Make'], ax=axs[1,1])
axs[1,1].set_title('Histogram of Car Make')
```

### Question 13

```python
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.show()
```

### Question 14

```python
plt.figure(figsize=(10, 6))
sns.lmplot(y='Odometer (KM)', x='Total Sales', hue='Make', data=car_sales, ci=None, markers='o', height=6)
plt.ylabel('Odometer (KM)')
plt.xlabel('Total Sales')
plt.show()
```

### Question 16

```python
correlation_matrix = car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

In terms of the correlation Total Sales and Price has a positive correlation which indicates as Price increases or decreases Total Sales will do the same. A moderate negative correlation can be shown with Price and Odomoeter which would show as the odomoeter is higher price would drop lower and vice versa. In a weak negative correlation total sales and odometer has a weaker negative correlation because sales wouldn't change as much with odometer unlike price with odometer.
